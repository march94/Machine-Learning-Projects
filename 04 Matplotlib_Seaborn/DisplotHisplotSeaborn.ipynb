```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C://Users//BRAVO15//Desktop//python//05-Seaborn/dm_office_sales.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(5,8))
sns.rugplot(x='salary',data=df, height=0.5)
```

```python
sns.set(style='darkgrid')
sns.displot(data=df,x='salary', bins=20, color='red', edgecolor='blue',lw=5, kde=True,rug=True)
```

```python
sns.histplot(data=df,x='salary', bins=20, color='red', edgecolor='blue',lw=5)
```

```python
sns.kdeplot(data=df,x='salary')
```
