```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a= np.linspace(0,10,11)
b=a**4
```

```python
b
```

```python
x = np.arange(0,10)
y=2*x
```

```python
y
```

```python
fig = plt.figure()
axes = fig.add_axes(rect=[0,0,1,1])
axes.plot(x,y)
plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_axes(rect=[0,0,1,1])
axes1.plot(a,b)
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

axes2 = fig.add_axes(rect=[0.2,0.5,0.25,0.25])
axes2.plot(a,b)
axes2.set_xlim(1,2)
axes2.set_ylim(0,30)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoom IN')
plt.show()
```

```python
fig = plt.figure(figsize=(2,2),dpi=100) #figsize inches
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
fig.savefig('new_figure.png', bbox_inches='tight')
```
