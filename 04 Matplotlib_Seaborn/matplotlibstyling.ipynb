```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color='#3ca832', label= 'x vs x',lw=10,ls=':') #hex color picker, can call linestyle as well if needed
ax.plot(x,x**2,'r--', label='x vs x^2',linewidth=5) 
ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
lines= ax.plot(x,x**2,'r', label='x vs x^2',linewidth=1,marker='o',ms=10, markerfacecolor='orange', markeredgewidth=3, markeredgecolor='r') # or call markersize
lines[0].set_dashes([20,20,100,20])
ax.legend()
```
