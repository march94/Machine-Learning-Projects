```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a= np.linspace(0,10,11)
b=a**4
```

```python
x=np.arange(0,10)
y=2*x
```

```python
fig, axes= plt.subplots(nrows=2, ncols=2)
axes[0][0].plot(x,y)
axes[1][1].plot(a,b)
```

```python
fig, axes= plt.subplots(nrows=3, ncols=1)
#axes[0].plot(x,y)
#axes[1].plot(a,b)

for x in axes:
    x.plot(a,b)

plt.tight_layout()
```

```python
fig, axes= plt.subplots(nrows=2, ncols=2, figsize=(10,10), dpi=200)
axes[0][0].plot(x,y)
axes[0][1].plot(x,y)
axes[1][0].plot(x,y)
axes[1][1].plot(a,b)
axes[1][1].set_xlabel('x label')
axes[1][1].set_title('title')
#axes[1].plot(a,b)
fig.suptitle("Figure Level Title", fontsize=16)
fig.subplots_adjust(wspace=0.5,hspace=0.5) #fraction of axes
fig.savefig("new_subplots.png",bbox_inches='tight')
```
