```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C://Users//BRAVO15//Desktop//python//05-Seaborn/StudentsPerformance.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df,y='math score', x='parental level of education', hue = 'test preparation course')
plt.legend(loc=(1.1,0.5))
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df,x='math score', y='parental level of education', hue = 'test preparation course', palette='Set2')
plt.legend(loc=(1.1,0.5))
```

```python
plt.figure(figsize=(10,5))
sns.violinplot(data=df,x='math score', y='parental level of education', inner='quartile')
```

```python
plt.figure(figsize=(10,5), dpi=200)
sns.swarmplot(data=df,x='math score',y='gender', hue= 'test preparation course',size=2, palette= 'Set2', dodge=True)
```

```python
sns.boxenplot(data=df, x='math score', y= 'test preparation course', hue='gender')
```
