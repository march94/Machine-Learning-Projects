```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./DATA/cluster_mpg.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df['origin'].value_counts()
```

```python
df_w_dummies=pd.get_dummies(df.drop('name', axis=1))
```

```python
df_w_dummies.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler= MinMaxScaler()
```

```python
scaled_X= scaler.fit_transform(df_w_dummies)
```

```python
scaled_X
```

```python
scaled_df= pd.DataFrame(scaled_X, columns=df_w_dummies.columns)
```

```python
scaled_df
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(scaled_df)
```

```python
plt.figure(figsize=(15,8))
sns.clustermap(scaled_df, col_cluster=False)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model =AgglomerativeClustering(n_clusters=4)
```

```python
cluster_label=model.fit_predict(scaled_df)
```

```python
cluster_label
```

```python
sns.scatterplot(data=df, x='mpg', y='weight', hue=cluster_label)
```

```python
np.sqrt(len(scaled_df.columns)) #max distance
```

```python
model = AgglomerativeClustering(n_clusters=None, distance_threshold=0) # 0 distance threshold, even if the distance metric is 0, also dont merge into a cluster
```

```python
model.fit_predict(scaled_df)
```

```python
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster import hierarchy
```

```python
model.children_
```

```python
linkage_matrix = hierarchy.linkage(model.children_)
```

```python
linkage_matrix
```

```python
plt.figure(figsize=(20,10))
dendro = dendrogram(linkage_matrix)
```

```python
plt.figure(figsize=(20,10))
dendro = dendrogram(linkage_matrix, truncate_mode='lastp',p=10)
```

```python
plt.figure(figsize=(20,10))
dendro = dendrogram(linkage_matrix, truncate_mode='level',p=5)
```
